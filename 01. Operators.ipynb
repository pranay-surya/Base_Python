## Python Operators 
    An operator is a special symbol or keyword that performs an operation on one or more values (called operands) and returns a result.

### 1. Arithmetic Operators
    Arithmetic operators Perform mathematical calculations. Used to carry out basic math operations such as addition, subtraction, multiplication, division, modulus, exponentiation, and floor division.

```python
a = 10
b = 3
print("Addition:", a + b)
print("Subtraction:", a - b)
print("Multiplication:", a * b)
print("Division:", a / b)
print("Floor Division:", a // b)
print("Modulus:", a % b)
print("Exponentiation:", a ** b)
```

### 2. Assignment Operators
    Assignment operators assign and update variable values. Used to assign values to variables and update them using shorthand operators.

```python
x = 5
print("Before Any Operation x ::",x )
x += 3
print("After += :", x)
x -= 2
print("After -= :", x)
x *= 2
print("After *= :", x)
x /= 4
print("After /= :", x)
x %= 2
print("After %= :", x)
```

### 3. Comparison Operators
    Comparison operators Compare values and return True or False. Used to compare two operands and produce a Boolean result (True or False) based on their relationship.

```python
print("a == b:", a == b)
print("a != b:", a != b)
print("a > b:", a > b)
print("a < b:", a < b)
print("a >= b:", a >= b)
print("a <= b:", a <= b)
```

### 4. Logical Operators
    Logical operators combine or modify conditional expressions. Used to combine multiple conditions or reverse a condition, mainly in decision-making statements.

```python
x = 10
print("Logical AND:", x > 5 and x < 20)
print("Logical OR:", x < 5 or x < 20)
print("Logical NOT:", not(x > 5))
```

### 5. Identity Operators
    Identity operators check whether two variables refer to the same object in memory. Used to determine if two variables point to the same memory location, not just equal values.

```python
y = [1, 2, 3]
z = y
w = [1, 2, 3]
print("y is z:", y is z)
print("y is w:", y is w)
print("y is not w:", y is not w)
```

### 6. Membership Operators
    Membership operators check the presence of a value in a sequence. Used to test whether a value exists in a sequence such as a list, tuple, string, or set.

```python
fruits = ["apple", "banana", "cherry"]
print("'banana' in fruits:", "banana" in fruits)
print("'grape' not in fruits:", "grape" not in fruits)
```

### 7. Bitwise Operators
    Bitwise operators perform operations at the binary (bit) level. Used to manipulate individual bits of integer values by applying binary logic.

```python
p = 5   # binary: 0101
q = 3   # binary: 0011
print("Bitwise AND:", p & q)
print("Bitwise OR:", p | q)
print("Bitwise XOR:", p ^ q)
print("Bitwise NOT:", ~p)
print("Bitwise Left Shift:", p << 1)
print("Bitwise Right Shift:", p >> 1)
```
